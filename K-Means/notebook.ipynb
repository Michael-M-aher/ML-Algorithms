# K-Means Clustering

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

from k_means import KMeans
```

### Generating the Data

```python
x, y = make_blobs(n_samples=100, centers=3, n_features=2, random_state=0)
```

### Plotting the Data

```python
plt.scatter(x[:, 0], x[:, 1], c=y, s=40, cmap='viridis')
plt.show()
```

### K-Means Clustering Model

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(x)
```

### Testing the Model

```python
y_pred = kmeans.predict(x)
accuracy = kmeans.accuracy(y, y_pred)
print('Accuracy: ', accuracy)
```

```python
# test the algorithm
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

ax1.scatter(x[:, 0], x[:, 1], c=y)
ax1.set_title('Original')

ax2.scatter(x[:, 0], x[:, 1], c=kmeans.predict(x))
ax2.scatter(kmeans.centroids[0][0], kmeans.centroids[0][1], c='r', marker='x')
ax2.scatter(kmeans.centroids[1][0], kmeans.centroids[1][1], c='r', marker='x')
ax2.scatter(kmeans.centroids[2][0], kmeans.centroids[2][1], c='r', marker='x')
ax2.set_title('KMeans')

plt.show()
```
